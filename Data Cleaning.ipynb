```python
import pandas as pd
import datetime
import json
import string

from nltk.corpus import stopwords
```

```python
# US data into dataframe
df = pd.read_csv('./output/USvideos.csv') 
df
```

```python
# Remove unecessary columns
df = df.drop(['thumbnail_link', 'comments_disabled', 'ratings_disabled'], axis=1)
df.head()
```

```python
# Reformate published date
df['publish_time'] = df['publish_time'].apply(lambda x: x[2:10])
df['publish_time'] = df['publish_time'].apply(lambda x: str(x).replace('-', '.'))
df.head()
```

```python
# Reformat trending date
df['trending_date'] = df['trending_date'].apply(lambda x: datetime.datetime.strptime(str(x), '%y.%d.%m').strftime('%y.%m.%d'))
df['trending_date']
```

```python
# Split tags
df['tags'] = df['tags'].apply(lambda x: str(x).split('|'))
df
```

```python
# Get category by number
cat_dict = {}

with open('./output/US_category_id.json') as f:
    data = json.loads(f.read())
    for i in range(0, 32):
        cat_dict[data['items'][i]['id']] = data['items'][i]['snippet']['title']
    
    
 
# for i in range(0, 32):
cat_dict
```

```python
# Convert category number to category name
df['category_id'] = df['category_id'].apply(lambda x: cat_dict[str(x)])
df
```

```python
df_dataset = pd.read_csv('./output/US_cleaned_data_2018.csv')
df_dataset.head()
```

```python
# https://adataanalyst.com/scikit-learn/countvectorizer-sklearn-example/
def text_process(mess):
    """
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    3. Return String
    """
    # Check characters to see if they are in punctuation
    nopunc = [char for char in mess if char not in string.punctuation]

    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)
    
    # Now just remove any stopwords
    list_words = nopunc.split()
    
    sentence = ' '.join(list_words)
    
    return sentence.lower()

def rem_stop_words(mess):

    list_words = [word for word in mess.split() if word not in stopwords.words('english')]
    return ' '.join(list_words)
```

```python
df_titles_info = pd.DataFrame(df_dataset['title'].dropna())

df_titles_info['filter_title'] = df_titles_info['title'].apply(text_process)
df_titles_info['filter_title_no_stops'] = df_titles_info['filter_title'].apply(rem_stop_words)

df_titles_info['category_id'] = df_dataset['category_id']

df_titles_info.to_csv('./output/US_count_vectorizer_dataset.csv', index=False)

df_titles_info
```
